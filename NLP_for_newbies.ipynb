## Step 1. Library Import & Data Load

````python
import pandas as pd 
import numpy as np 
````

## Step 2. Data Preprocessing
### 2-a. Drop Columns

````python
train_df.head()
````

````python
train_df.drop(columns=['id','keyword','location'], axis=1, inplace=True)
````

````python
test_df.head()
````

````python
test_df.drop(columns=['keyword','location'],axis=1, inplace=True)
````

````python
print(train_df.shape, test_df.shape)
````

### 2-b. Tokenizer

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_valid, y_train, y_valid = train_test_split(train_df['text'], train_df['target'], test_size=0.2, random_state=42)
````

````python
print(X_train.shape, y_train.shape, X_valid.shape, y_valid.shape)
````

````python
from tensorflow.keras.preprocessing.text import Tokenizer
````

````python
vocab_size = 1000
oov_token = "<OOV>"
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
````

````python
tokenizer.fit_on_texts(X_train)
````

````python
X_train = tokenizer.texts_to_sequences(X_train)
X_valid = tokenizer.texts_to_sequences(X_valid)
````

````python
for i in range(10):
    print(len(X_train[i]))
````

````python
X_train[0]
````

````python
for i in range(10):
    print(len(X_valid[i]))
````

````python
X_valid[0]
````

### 2-c. Pad Sequences

````python
from tensorflow.keras.preprocessing.sequence import pad_sequences
````

````python
max_length = 120
trunc_type = 'post'
pad_type = 'post'
````

````python
X_train_padded = pad_sequences(X_train, maxlen=max_length, truncating=trunc_type, padding=pad_type)
X_valid_padded = pad_sequences(X_valid, maxlen=max_length, truncating=trunc_type, padding=pad_type)
````

````python
X_train_padded[:2]
````

````python
X_valid_padded[:2]
````

````python
print(X_train_padded.shape, X_valid_padded.shape)
````

### 2-d. Match Data type to numpy.ndarray

````python
print(type(X_train_padded), type(X_valid_padded))
print(type(y_train), type(y_valid))
````

````python
y_train = np.array(y_train)
y_valid = np.array(y_valid)
````

````python
print(type(X_train_padded), type(X_valid_padded))
print(type(y_train), type(y_valid))
````

## Step 3. Modeling

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional, Flatten
````

````python
embedding_dim = 16
# vocab_size = 1000
# max_length = 120
````

````python
model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),
    Bidirectional(LSTM(64, return_sequences=True)),
    Bidirectional(LSTM(64, return_sequences=True)),
    Bidirectional(LSTM(64, dropout=0.5)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
````

````python
model.summary()
````

## Step 4. Model Compile

````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
````

## Step 5. Callbacks

````python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
````

````python
filepath = 'my_checkpoint.ckpt'
cp = ModelCheckpoint(
    filepath=filepath,
    save_weights_only=True,
    save_best_only=True,
    monitor='val_loss',
    verbose=1
)
````

````python
ep = EarlyStopping(
    monitor='val_loss', 
    patience=5,
)
````

## Step 6. Model Fit

````python
epochs=30
model.fit(
    X_train_padded, y_train,
    validation_data = (X_valid_padded, y_valid),
    callbacks=[cp,ep],
    epochs=epochs
)
````

## Step 7. Model Evaluate & Save

````python
model.load_weights(filepath)
````

````python
model.evaluate(X_valid_padded, y_valid)
````

````python
X_valid[0]
````

````python
model.save('./model/basic_nlp.h5')
````
